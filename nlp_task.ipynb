```python
from nltk.corpus import stopwords
import string
from stemming.porter2 import stem
 
# load doc into memory
def load_doc(filename):

    file = open(filename, 'r')
# read all text
    text = file.read()
    # close the file
    file.close()
    return text
 

def clean_doc(doc):

    tokens = doc.split()

    table = str.maketrans('', '', string.punctuation)
    tokens = [w.translate(table) for w in tokens]

    tokens = [word for word in tokens if word.isalpha()] #crucial part as our doc contains many alphanumeric and escape characters

    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]

    tokens = [word for word in tokens if len(word) > 1]
    return tokens

def stemmit(word_list):
    stem_list=[]
    for i in word_list:
        stem_list.append(stem(i))
    return stem_list




def yellow(word):
    positivity=0
    negativity=0
    word_count=0
    for i in word:
        word_count+=1
        if i in pos:
            positivity+=1
        elif i in neg:
            negativity+=1
    popularity_score=(positivity-negativity)/((positivity+negativity)+0.000001)  
    pos_prop=positivity/word_count
    neg_prop=negativity/word_count
    
    
    return (positivity,negativity,popularity_score,word_count,pos_prop,neg_prop)
    

    
def blue(txt,token):
    file1 = open(txt, 'r')
    Lines = file1.readlines()
    total_sent=len(Lines)
    words=0
    comp=0
    
    for i in stemit:
        words+=1
        if type(i)==str:
            v=dic.inserted(i).split('-')
            if len(v)>2:
                comp+=1
    avg_sentence_length= words/len(Lines)
    percentage_of_comp=comp/words
    fog_index=0.4 * (avg_sentence_length+percentage_of_comp )
    return(avg_sentence_length,percentage_of_comp,fog_index,comp)
        
def green(token):
    word=0
    costn=0
    unct=0
    for i in token:
        word+=1
        if i in constrain:
            costn+=1
        elif i in uncert:
            unct+=1
    if unct!=0 and costn !=0:
        unn_prop= unct/unct+costn
        cnn_prop=costn/unct+costn
        cnn_whole=costn/word
    else:
        unct=0
        costn=0
        unn_prop=0
        cnn_prop=0
        cnn_whole=0
    return(unct,costn,unn_prop,cnn_prop,cnn_whole)
            
    

file_name=[]
clean_txt=[]
stemmed=[]

# load the document

yello=[]
blu=[]
gre=[]

# final list
positive_score=[]
negative_score=[]
popularity_score=[]
word_count=[]
pos_word_prop=[]
neg_word_prop=[]

avg_sent_length=[]
perct_comp_words=[]
fog_index=[]
comp_word_count=[]

un_score=[]
co_score=[]
un_prop=[]
co_prop=[]
co_whole=[]


import os
for dirname, _, filenames in os.walk('./'):
    for filename_m in filenames:
        filename=os.path.join(dirname, filename_m)
        text = load_doc(filename)
        tokens = clean_doc(text)
        print(f'filename :{filename} tokenized sucessfully')
        stemit=stemmit(tokens)
        #print(tokens)
        file_name.append(filename)
        clean_txt.append(tokens)
        stemmed.append(stemit)     
        
        # to be displayed
        y=yellow(stemit)
        b=blue(filename,tokens)
        g=green(tokens)
        yello.append(y)
        blu.append(b)
        gre.append(g)
        
        # add to final list seperately
        
        positive_score.append((y[0]))
        negative_score.append(y[1])
        popularity_score.append(y[2])
        pos_word_prop.append(y[3])
        neg_word_prop.append(y[4])
        word_count.append(y[5])
        

        avg_sent_length.append(b[0])
        perct_comp_words.append(b[1])
        fog_index.append(b[2])
        comp_word_count.append(b[3])

        un_score.append(g[0])
        co_score.append(g[1])
        un_prop.append(g[2])
        co_prop.append(g[3])
        co_whole.append(g[4])
```

```python
maindf=pd.DataFrame(list(zip( 
        file_name,
        positive_score,
        negative_score,
        popularity_score,
        pos_word_prop,
        neg_word_prop,
        word_count,
        

        avg_sent_length,
        perct_comp_words,
        fog_index,
        comp_word_count,

        un_score,
        co_score,
        un_prop,
        co_prop,
        co_whole
)),columns=[ 'fileName','positive_score',
        'negative_score',
        'popularity_score',
        'pos_word_prop',
        'neg_word_prop',
        'word_count',
        

        'avg_sent_length',
        'perct_comp_words',
        'fog_index',
        'comp_word_count',

        'un_score',
        'co_score',
        'un_prop',
        'co_prop',
        'co_whole'])
```
